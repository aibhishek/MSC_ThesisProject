### Library Import

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
temp = temp.drop(['J-D','D-N','MAM','JJA','SON','DJF'], axis=1)
```

### Time Series Analysis using ARIMA

#### Getting Autocorrelation and differentiating

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
# Original Series
fig, axes = plt.subplots(3, 2, sharex=True)
axes[0, 0].plot(time_series_df.Total_Cloud_Occurence); axes[0, 0].set_title('Original Series')
plot_acf(time_series_df.Total_Cloud_Occurence, ax=axes[0, 1])

# 1st Differencing
axes[1, 0].plot(time_series_df.Total_Cloud_Occurence.diff()); axes[1, 0].set_title('1st Order Differencing')
plot_acf(time_series_df.Total_Cloud_Occurence.diff().dropna(), ax=axes[1, 1])

# 2nd Differencing
axes[2, 0].plot(time_series_df.Total_Cloud_Occurence.diff().diff()); axes[2, 0].set_title('2nd Order Differencing')
plot_acf(time_series_df.Total_Cloud_Occurence.diff().diff().dropna(), ax=axes[2, 1])
```

#### Building the ARIMA model

```python
from statsmodels.tsa.arima_model import ARIMA

# 1,1,2 ARIMA Model
model = ARIMA(time_series_df.Total_Cloud_Occurence, order=(4,1,4))
model_fit = model.fit(disp=0)
print(model_fit.summary())
```

#### Model without MA2

```python
# 4,1,4 ARIMA Model
model = ARIMA(time_series_df.Total_Cloud_Occurence, order=(4,1,4))
model_fit = model.fit(disp=0)
print(model_fit.summary())
```

#### The P>|z| value for ma.L1.D.Total_Cloud_Occurence is 0.00 which is less than 0.05. So we will use this model.

#### Plot residual errors

```python
residuals = pd.DataFrame(model_fit.resid)
fig, ax = plt.subplots(1,2)
residuals.plot(title="Residuals", ax=ax[0])
residuals.plot(kind='kde', title='Density', ax=ax[1])
plt.show()
```

#### The residual mean is near 0 and density curve is almost normally distributed. So we will proceed with predictions.

```python
model_fit.plot_predict(dynamic=False)
plt.show()
```

#### Creating test and train data

```python
train = time_series_df.Total_Cloud_Occurence[:100]
test = time_series_df.Total_Cloud_Occurence[100:]
```

```python
# Build Model
#model = ARIMA(train, order=(3,2,1))  
model = ARIMA(train, order=(4, 1, 4))  
fitted = model.fit(disp=-1)  

# Forecast
fc, se, conf = fitted.forecast(20, alpha=0.05)  # 95% conf

# Make as pandas series
fc_series = pd.Series(fc, index=test.index)
lower_series = pd.Series(conf[:, 0], index=test.index)
upper_series = pd.Series(conf[:, 1], index=test.index)

# Plot
plt.figure(figsize=(12,5), dpi=100)
plt.plot(train, label='training')
plt.plot(test, label='actual')
plt.plot(fc_series, label='forecast')
plt.fill_between(lower_series.index, lower_series, upper_series, 
                 color='k', alpha=.15)
plt.title('Forecast vs Actuals')
plt.legend(loc='upper left', fontsize=8)
plt.show()
```

```python
# Forecast
n_periods = 120
fc, se, confint = fitted.forecast(120, alpha=0.05)  # 95% conf
index_of_fc = np.arange(len(time_series_df.Total_Cloud_Occurence), len(time_series_df.Total_Cloud_Occurence)+n_periods)

# make series for plotting purpose
fc_series = pd.Series(fc, index=index_of_fc)
lower_series = pd.Series(confint[:, 0], index=index_of_fc)
upper_series = pd.Series(confint[:, 1], index=index_of_fc)

# Plot
plt.plot(time_series_df.Total_Cloud_Occurence)
plt.plot(fc_series, color='darkgreen')
plt.fill_between(lower_series.index, 
                 lower_series, 
                 upper_series, 
                 color='k', alpha=.15)

plt.title("Final Forecast of Total Cloud Occurence Usage")

plt.xlabel('Months')
plt.ylabel('Total Cloud Occurence')
plt.show()
```
